Write some code to compute the image gradient: $\sqrt{f_x^2+f_y^2}$

```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = 1.0 * plt.imread("assets/al.jpg") # convert to double
filt = np.array([[-0.1250, -0.2500, -0.1250], [0, 0, 0], [0.1250, 0.2500, 0.1250]])
```

```python
img_x = ndimage.convolve(img, filt, mode='reflect')
img_y = ndimage.convolve(img, filt.T, mode='reflect')
img_g = np.sqrt(img_x**2 + img_y**2)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img_y, cmap='gray')
plt.subplot(1, 3, 2)
plt.imshow(img_x, cmap='gray')
plt.subplot(1, 3, 3)
plt.imshow(img_g, cmap='gray')
```
