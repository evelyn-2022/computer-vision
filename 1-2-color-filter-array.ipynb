Write some code that generates a 256 x 256 image. The value of each pixel in the $i^{th}$ image row of the red channel is $i*0.0039$, the value of each pixel in the $i^{th}$ image row of the green channel is $1-i*0.0039$, and every pixel in the blue channel is 1. We will encode pixel values in the range [0,1] and display the image using `matplotlib.pyplot.imshow`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 256
im = np.zeros((N, N, 3))
for i in range(N):
    for j in range(N):
        im[i, j, 0] = i * 0.0039
        im[i, j, 1] = 1 - i * 0.0039
        im[i, j, 2] = 1
```

```python
# plt.axis("off")
plt.imshow(im)
```
